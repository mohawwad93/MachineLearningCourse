## 1 - Packages

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 2 - Overview of the Problem set

**Problem Statement**: You are given the iris dataset which consists of 3 different types of irises that are Setosa, Versicolour, and Virginica. The petal and sepal length and width are stored in a 150x4 numpy array.
Thus, The dataset contains 150 iris sample where each sample has four features that are the petal and sepal length and width.

We will build a simple gaussian naive bayes algorithm that can correctly classify irises as Setosa, Versicolour, or Virginica.

Let's get more familiar with the dataset. Load the data by running the following code.

```python
# Loading iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# Explore the dataset
print("Name   of the features: ", iris.feature_names)
print("Name   of the target  : ", iris.target_names)
print("Shape  of the dataset : ", iris.data.shape)
print("Number of the samples : ", iris.data.shape[0])
print("Number of the features: ", iris.data.shape[1])
print("Number of the targets : ", iris.target.shape)
```

```python
# Display a portion of the dataset
target_col_name = 'target'
df = pd.DataFrame(data= np.hstack([iris.data, iris.target[:, np.newaxis]]),
                       columns= iris.feature_names + [target_col_name])
df.head()
```

Note: You should reshape iris target shape from (150,) to (150,1) to treat it as column vector.

## 3 - Splitting Data into Training and Test Sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df[iris.feature_names], df[target_col_name], train_size=0.6, random_state=0)

print(f"Number of training examples: X_train = {X_train.shape[0]}")
print(f"Number of testing examples: X_test = {X_test.shape[0]}")
print(f"X_train shape: {X_train.shape}")
print(f"Y_train shape: {Y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Y_test shape: {Y_test.shape}")
```

## 4 - Building Model


```python
# Create a classifer
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
```

```python
# Train the classifer
clf.fit(X_train, Y_train)
```

## 5 - Measuring Model Performance

```python
# Predict unseen irises
clf.predict(X_test)
```

```python
# Actual unseen irises
Y_test.values
```

```python
train_accuracy = clf.score(X_train, Y_train)
test_accuracy = clf.score(X_test, Y_test)
print('Train Accuracy: ', train_accuracy)
print('Test Accuracy: ', test_accuracy)
```

## 6 - Displaying Gaussian Distribution

```python
import seaborn as sns
sns.set()

def plot_decision_boundary(model, df):
    # Set classes color
    classes_color = ('blue','orange','green')
    # Set min and max values and give it some padding
    xx_min, xx_max  = df[iris.feature_names[-2]].min()-0.5, df[iris.feature_names[-2]].max()+0.5
    yy_min, yy_max = df[iris.feature_names[-1]].min()-0.5, df[iris.feature_names[-1]].max()+0.5
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(xx_min, xx_max, h), np.arange(yy_min, yy_max, h))
    # Computing the predicted class function for each value on the grid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel(), xx.ravel(), yy.ravel()])
    # Reshaping the predicted class into the meshgrid shape
    Z = Z.reshape(xx.shape)
    # Select required columns to display
    data = df.iloc[:, -3:]
    # Plot the contour and training examples
    grid = sns.FacetGrid(data, hue=data.columns[-1], height=11, palette='colorblind'
                        ).map(plt.scatter,iris.feature_names[-2],iris.feature_names[-1]).add_legend()
    my_ax = grid.ax
    # Plot the filled and boundary contours
    my_ax.contourf(xx, yy, Z, 2, alpha = 0.1, colors = classes_color)
    my_ax.contour(xx, yy, Z, 2, alpha = 1, colors = classes_color)
    # Add axis labels and title
    my_ax.set_xlabel(iris.feature_names[-2])
    my_ax.set_ylabel(iris.feature_names[-1])
    my_ax.set_title('Gaussian Naive Bayes Decision Boundaries')
    
    # Plot Gaussian distribution
    xg = np.linspace(xx_min, xx_max, 60)
    yg = np.linspace(yy_min, yy_max, 40)
    grid = np.vstack([xx.ravel(), yy.ravel()]).T
    X = df.iloc[:, -3:-1].values
    Y = df.iloc[:, -1].values
    for label, color in enumerate(classes_color):
        mask = (Y == label)
        mu, std = X[mask].mean(0), X[mask].std(0)
        P = np.exp(-0.5 * (grid - mu) ** 2 / std ** 2).prod(1)
        Pm = np.ma.masked_array(P, P < 0.03)
        my_ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,cmap=color.title() + 's')
        my_ax.contour(xx, yy, P.reshape(xx.shape),levels=[0.01, 0.1, 0.5, 0.9], colors=color, alpha=0.2)
    # Show graph
    plt.show()
    
plot_decision_boundary(clf, df)
```

## 7 - Exploring Model Parameters

```python
print(f'Number of training samples observed in each class = {clf.class_count_}')
print(f'Probability of each class = {clf.class_prior_}')
print(f'Mean of each feature per class = {clf.theta_}')
print(f'Variance of each feature per class = {clf.sigma_}') 
```

```python
# Check model's parameters values
class_num = 3
feature_num = 4
for class_ in range(class_num):
    mask  = Y_train == class_
    print('*'*15, f"Class {class_}", '*'*15)
    for i in range(feature_num):
        feature_col = X_train.loc[mask, iris.feature_names[i]]
        print(f'{iris.feature_names[i]} mean : {feature_col.mean()}')
        print(f'{iris.feature_names[i]} variance : {feature_col.var()}')
```

**Note: there is a slight difference in variance values due to var_smoothing parameter in GaussianNB which used for calculation stability**

## 8 - Suggested Model

We will see how tree splitting dataset into train and test sets could effect model performance!

```python
def model(X, Y, seed_split):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y,train_size=0.7, random_state=seed_split)
    clf = GaussianNB()
    clf.fit(X_train, Y_train) 
    return clf

clf = model(df[iris.feature_names], df[target_col_name], seed_split=9)
train_accuracy = clf.score(X_train, Y_train)
test_accuracy = clf.score(X_test, Y_test)
print(f'Train Accuracy: {train_accuracy}')
print(f'Test Accuracy: {test_accuracy}')
plot_decision_boundary(clf, df)
```

**Note: GaussianNB doesn't have hyperparameters to tune which makes GaussianNB good simple fast classifier for quick results.**
